```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("C:\\Users\\Lauta\\Downloads\\dataset_finalparalaprueba.csv")
# Análisis Univariado
plt.figure(figsize=(10, 5))
sns.histplot(df['release_date'], bins=30, kde=True)
plt.title('Distribución de Años de Lanzamiento')
plt.xlabel('Año')
plt.ylabel('Cantidad de Películas')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("C:\\Users\\Lauta\\Downloads\\dataset_finalparalaprueba.csv")

# Películas por Mes
df['release_month'] = df['release_date']
movies_per_month = df['release_month'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
movies_per_month.plot(kind='bar')
plt.title('Películas Lanzadas por Mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Películas', rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("C:\\Users\\Lauta\\Downloads\\dataset_finalparalaprueba.csv")

# Películas por Día
df['release_day'] = df['release_date']
movies_per_day = df['release_day'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
movies_per_day.plot(kind='bar')
plt.title('Películas Lanzadas por Día')
plt.xlabel('Día')
plt.ylabel('Cantidad de Películas',rotation=45)
plt.show()
```
